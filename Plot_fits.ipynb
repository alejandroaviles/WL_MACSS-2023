```python
import matplotlib.pyplot as plt
import numpy as np
from astropy.visualization import astropy_mpl_style
plt.style.use(astropy_mpl_style)
from astropy.io import fits
from astropy.utils.data import get_pkg_data_filename
import os
```

```python
os.chdir('WL_MACSS-2023')
```

```python
image_file =get_pkg_data_filename('kappa_A.0.fits')
fits.info(image_file)
image_data = fits.getdata(image_file, ext=0)
```

```python
plt.figure()
plt.imshow(image_data,filternorm=False,interpolation='none')
plt.xticks([])
plt.yticks([])
plt.grid(False)
```

```python
gamma_file =get_pkg_data_filename('gamma_A.0.fits')
fits.info(gamma_file)
gamma_data = fits.getdata(gamma_file, ext=0)
```

```python
x=np.zeros(1024)
y=np.zeros(1024)

for i in range(1024):
    x[i]=i+1;
    y[i]=i+1;
```

```python
gamma1=gamma_data[:,:, 0]
gamma2=gamma_data[:,:, 1]
```

```python
print(gamma_data.shape)
print(gamma1.shape)
print(gamma2.shape)
```

```python
x1=np.zeros(1024*1024)
x2=np.zeros(1024*1024)
g1=np.zeros(1024*1024)
g2=np.zeros(1024*1024)

counter=0;
for i in range(1024):
    for j in range(1024):
        x1[counter]=i+1;
        x2[counter]=j+1;
        g1[counter]=gamma1[i,j];
        g2[counter]=gamma2[i,j];
        counter=counter+1;
```

```python
nnn=234567
print(g1[nnn],g2[nnn])
```

```python
div=16
n=int(1024/div)
x1=np.zeros(n*n)
x2=np.zeros(n*n)
g1=np.zeros(n*n)
g2=np.zeros(n*n)

counter=0;
for i in range(n):
    for j in range(n):
        x1[counter]=i*div+1;
        x2[counter]=j*div+1;
        g1[counter]=gamma1[i*div,j*div];
        g2[counter]=gamma2[i*div,j*div];
        counter=counter+1;
```

```python
quiveropts = dict(color='black', headlength=0, pivot='middle', scale=3, linewidth=.5, units='xy', width=.05, headwidth=0)
plt.quiver(x1, x2, g1, g2, color='white',headaxislength=0,width=0.0015)

plt.imshow(image_data,filternorm=False,interpolation='none')
plt.grid(False)
#plt.savefig("kappa_A0.jpg", format="jpg", bbox_inches="tight")

plt.xticks([])
plt.yticks([])
plt.show()
```
